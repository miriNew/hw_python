# גרף פיזור

```python
x = np.random.rand(50) # 50 ערכי x אקראיים
y = np.random.rand(50) # 50 ערכי y אקראיים
colors = np.random.rand(50) # 50 צבעים אקראיים
sizes = 1000 * np.random.rand(50) # 50 גדלים אקראיים לנקודות
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # הצגת סרגל הצבעים
plt.title('Scatter Plot Example')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

# גרף עם שני ייצוגים שונים

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, '-b', label='sin(x)')
plt.plot(x, y2, ':r', label='cos(x)')
plt.scatter(x, y1, c='blue', s=10)  # נקודות לסינוס
plt.title('Line and Scatter on the Same Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```

# subplots

```python
fig, axs = plt.subplots(2, 2)  # יוצר מטריצה של גרפים בגודל 2x2
axs[0, 0].plot(x, y1, 'tab:blue')
axs[0, 0].set_title('Sin(x)')
axs[0, 1].plot(x, y2, 'tab:red')
axs[0, 1].set_title('Cos(x)')
axs[1, 0].scatter(x, y1, c='yellow')
axs[1, 0].set_title('Sin(x) Scatter')
axs[1, 1].scatter(x, y2, c='green')
axs[1, 1].set_title('Cos(x) Scatter')

# התאמת הגרפים למראה נקי יותר
for ax in axs.flat:
    ax.label_outer()
```

```python
 years = range(2010, 2021)
 consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
 consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
 consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))  # יוצר מטריצה של גרפים בגודל 2x2
axs[0, 0].plot(years, consumption_oil, color='brown', marker='o')
axs[0, 0].set_title('צריכת שמן')
axs[0, 0].set_ylabel('יחידות')

axs[0, 1].plot(years, consumption_gas, color='gray', marker='s')
axs[0, 1].set_title('צריכת גז')

axs[1, 0].plot(years, consumption_renewable, color='green', marker='^')
axs[1, 0].set_title('אנרגיה מתחדשת')
axs[1, 0].set_xlabel('שנה')
axs[1, 0].set_ylabel('יחידות')

# גרף 4: השוואה כוללת
axs[1, 1].plot(years, consumption_oil, label='שמן', color='brown')
axs[1, 1].plot(years, consumption_gas, label='גז', color='gray')
axs[1, 1].plot(years, consumption_renewable, label='מתחדשת', color='green')
axs[1, 1].set_title('השוואה כוללת')
axs[1, 1].legend()



# התאמת הגרפים למראה נקי יותר
for ax in axs.flat:
    ax.label_outer()
```
